```python
import numpy as np

# Example: First row ke 'AI Adoption Rate (%)' column ko null banao
df.loc[0, 'AI Adoption Rate (%)'] = np.nan
# Randomly 5 rows me 'Job Loss Due to AI (%)' ko NaN banao
df.loc[df.sample(5).index, 'Job Loss Due to AI (%)'] = np.nan
import random

# Randomly 20% rows me 'Revenue Increase Due to AI (%)' ko NaN karo
n = int(0.2 * len(df))
indices = random.sample(range(len(df)), n)
df.loc[indices, 'Revenue Increase Due to AI (%)'] = np.nan
```

```python
df.isnull().sum()
```

**Step 3:Handling missing values**

```python
df['AI Adoption Rate (%)'].median()
```

```python
#fill missing ai adoption rate with median
df['AI Adoption Rate (%)'].fillna(df['AI Adoption Rate (%)'].median(),inplace=True)
#fill missing 'job loss due to ai' with most frequent value
df['Job Loss Due to AI (%)'].fillna(df['Job Loss Due to AI (%)'].mode()[0],inplace=True)
#drop 'revenue increase due to ai (%)' columns(to many missing value)
df.drop(columns=['Revenue Increase Due to AI (%)'],inplace=True)
```

# **🧾 Dataset Description: Global AI Content Impact Dataset**
This dataset provides a comprehensive overview of how Artificial Intelligence (AI) is influencing industries and countries worldwide through content generation, adoption, and public perception. It captures both quantitative and qualitative metrics across a range of sectors and regions, offering valuable insights for analysts, policymakers, and researchers.


# **📁 Key Attributes (Columns)**
Column Name	Description
Country	The country where the data point was recorded.
Industry	The sector or industry being evaluated (e.g., Education, Finance, Gaming, etc.).
AI Adoption Rate (%)	The percentage of companies in that industry/country using AI technologies.
AI Content Volume (TB/year)	The estimated amount of AI-generated content produced annually, measured in terabytes.
Job Loss Due to AI (%)	The estimated percentage of job displacement attributed to AI tools.
Revenue Increase (%)	The reported or projected increase in revenue due to AI implementation.
Human-AI Collaboration Level (%)	The extent to which humans and AI systems work together (e.g., co-editing, decision support).
Consumer Trust in AI (%)	A survey-based score indicating public trust in AI-generated or AI-assisted content.
AI Companies' Market Share (%)	The percentage of market share held by AI-first companies in that sector.
Most Used AI Tool	The dominant AI tool used (e.g., ChatGPT, Midjourney, DALL·E, etc.).


**📌 Dataset Highlights**   
Covers multiple countries and industries for a global perspective.

Explores the trade-off between automation and employment.

Highlights trends in consumer trust and AI-human collaboration.

Helps evaluate the market impact of AI content creators and tools.
